```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
facecolour = '#FFFFFF'
axiscolour = '#000000'
secondarycolour = '#008000'
tertiarycolour = '#ff0000'

def color_axis(ax, facecolour=facecolour, axiscolour=axiscolour):
    ax.spines['bottom'].set_color(axiscolour)
    ax.spines['top'].set_color(facecolour)
    ax.spines['left'].set_color(axiscolour)
    ax.spines['right'].set_color(facecolour)
    ax.tick_params(axis='x',colors=axiscolour)
    ax.tick_params(axis='y',colors=axiscolour)
    ax.xaxis.label.set_color(axiscolour)
    ax.yaxis.label.set_color(axiscolour)
    return ax
```

```python
sns.set(style='white',
        rc={'axes.facecolor':facecolour, 
            'figure.facecolor':facecolour})
```

```python
data=pd.read_csv('../part-01/complete_data.csv')
```

```python
p=data[['total_points','player']].groupby('player').mean()
v=data[['player','value']].groupby('player').mean()
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.distplot(a=p,bins=30, kde=False, color=axiscolour,
            hist_kws={'alpha':1,'linewidth':0}, ax=ax)
sns.despine(offset=10,trim=True)
ax=color_axis(ax)
plt.savefig('point_histo.png', facecolor=facecolour)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax=sns.distplot(a=v,bins=30, kde=False, color=axiscolour,
            hist_kws={'alpha':1,'linewidth':0})
sns.despine(offset=10,trim=True)
ax=color_axis(ax)
plt.savefig('value_histo.png', facecolor=facecolour)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.regplot(x=data['total_points'],y=data['value'], ax=ax,
           color=axiscolour)
sns.despine(offset=10,trim=True)
ax=color_axis(ax)
plt.show()
```

```python
mean_pts=[]
mean_val=[]
for player in data['player'].unique():
    pts = data[data['player']==player]['total_points'].mean()
    val = data[data['player']==player]['value'].mean()
    mean_pts.append(pts)
    mean_val.append(val)
    
    
fig, ax = plt.subplots(figsize=(15,10))
sns.regplot(x=mean_pts,y=mean_val, ax=ax,
           color=axiscolour)
sns.despine(offset=10,trim=True)
ax=color_axis(ax)
plt.savefig('pVv.png', facecolor=facecolour)
```

```python
from statsmodels.tsa.stattools import acf, pacf, ccf

def lag_with_me(lag,me,data,start,end,diff,filename1,filename2):
    acf_dict={}
    for player in data['player'].unique():
        if diff==True:
            lag_list = data[(data['player']==player)&
                           (data['adj_round']>=start)&
                           (data['adj_round']<=end)][lag].diff().dropna().values
            me_list = data[(data['player']==player)&
                           (data['adj_round']>=start)&
                           (data['adj_round']<=end)][me].diff().dropna().values
        else:
            lag_list = data[(data['player']==player)&
                           (data['adj_round']>=start)&
                           (data['adj_round']<=end)][lag].dropna().values
            me_list = data[(data['player']==player)&
                           (data['adj_round']>=start)&
                           (data['adj_round']<=end)][me].dropna().values
        apps=data[(data['player']==player)&
                 (data['adj_round']>=start)&
                 (data['adj_round']<=end)]['adj_round'].values
        all_rounds_me=[0]*(end-start)
        all_rounds_lag=[0]*(end-start)
        for gw, lag_el, me_el in zip(apps,lag_list,me_list):
            all_rounds_me[gw-1-start]=me_el
            all_rounds_lag[gw-1-start]=lag_el
            
        all_rounds_me=np.asarray(all_rounds_me)
        all_rounds_lag=np.asarray(all_rounds_lag)
        acf_dict[player]=acf(ccf(all_rounds_me,all_rounds_lag),nlags=(end-start))
    
    df_acf=pd.DataFrame(acf_dict)
    
    best_lag={}
    for player in df_acf.columns:
        try:
            best_lag[player]=df_acf.loc[1:,player].abs().idxmax()
        except:
            continue
            
    best_lag=pd.Series(best_lag)
    
    
    fig, ax = plt.subplots(figsize=(15,10))
    ax=sns.distplot(a=best_lag,bins=30, kde=False, color=axiscolour,
                hist_kws={'alpha':1,'linewidth':0})
    ax=color_axis(ax)
    plt.savefig(f'{filename1}.png', facecolor=facecolour)
    
    fig, ax = plt.subplots(figsize=(15,10))
    ax=sns.barplot(list(df_acf.loc[1:,:].mean(axis=1).index),
                   df_acf.loc[1:,:].mean(axis=1),
                   color=axiscolour,linewidth=0, saturation=1)
    ax=color_axis(ax)
    plt.savefig(f'{filename2}.png', facecolor=facecolour)
```
